```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.southwest.com/flight/search-flight.html?preserveBugFareType=TRUE"
```

```python
roundTrip = 'true'
originAirport = 'DAL'
destinationAirport = 'STL'
outDate = '10/27/2016'
outTime = 'ANYTIME'
returnDate = '10/28/2016'
returnTime = 'ANYTIME'
adultCount = '1'
seniorCount = '0'
fareType = 'DOLLARS'
```

```python
form_data = {
    'twoWayTrip': roundTrip,
    #'originAirport_displayed': originAirport,
    #'destinationAirport_displayed': destinationAirport,
    'originAirport': originAirport,
    'destinationAirport': destinationAirport,
    'outboundDateString': outDate,
    'outboundTimeOfDay': outTime,
    'returnDateString': returnDate,
    'returnTimeOfDay': returnTime,
    'adultPassengerCount': adultCount,
    'seniorPassengerCount': seniorCount,
    'fareType': fareType,
    'submitButton': 'submit',
}
```

```python
response = requests.post(url, data=form_data)
```

```python
soup = BeautifulSoup(response.content, "lxml")
```

```python
outbound_table = soup.find('table', attrs={'id': 'faresOutbound'}).find('tbody')
```

```python
data = []
for row in outbound_table.find_all('tr'):
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append(cols)
```
